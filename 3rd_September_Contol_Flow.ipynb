11. Write a program to find the largest among three numbers using nested if-else statements.


```python
# Taking three number as a input---
a = int(input("Enter the first number : "))
b = int(input("Enter the second number : "))
c = int(input("Enter the third number : "))

# Using nested loop

if a >= b :
    if a > c:
        print(a," is largest among the three numbers")
    else:
        print(c," is largest among the three numbers")
else:
    if b >=c:
        print(b," is largest among the three numbers")
    else:
        print(c," is largest among the three numbers")
```

12. Implement a program to determine if a triangle is equilateral, isosceles, or scalene.


```python
# Taking lengths of the side of the triangle as an input---->
s1 = int(input("Enter the first length of side: "))
s2 = int(input("Enter the second length of side: "))
s3 = int(input("Enter the third length of side: "))

if s1 == s2 == s3:
    print("The triangle is equilateral triangle ")
else:
    if s1 == s2 or s2 == s3 or s3 == s1:
        print("The triangle is isosceles triangle ")
    else:
        print("The triangle is scalene triangle ")
```

20. Develop a program that checks if a given number is prime or not using nested if-else statements.

```python
num = int(input("Enter a number: "))

if num<2:
    print(num ,"is not a prime number")
else:
    prime = True
    for i in range(2,num):
        if num%i==0:
            prime = False
            break
    if prime:
        print(num ,"is a prime number")
    else:
        print(num ,"is not a prime number")
```
